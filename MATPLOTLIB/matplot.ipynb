```python

import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np 
x=np.linspace(0,1,20)
y=x**4
#functional method
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot 1')
plt.subplot(1,3,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np 
x=np.linspace(0,1,20)
y=x**2
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plot 2')
```
