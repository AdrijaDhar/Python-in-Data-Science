```python

import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np 
x=np.linspace(0,1,20)
y=x**4
#functional method
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot 1')
plt.subplot(1,3,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(x,y,'r')

```

```python
import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np 
x=np.linspace(0,1,20)
y=x**2
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plot 2')
```

```python
import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np
fig=plt.figure()
x=np.linspace(0,1,20)
y=x**3
axes1=fig.add_axes([0.1,0.1,.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python

import matplotlib.pyplot as plt 
#%matplotlib online
import numpy as np
fig=plt.figure()
x=np.linspace(0,1,20)
y=x**3
fig,axes=plt.subplots(nrows=1,ncols=2)

for curr_ax in axes:
    curr_ax.plot(x,y,label='Cube root')
    x,y=y,x
plt.tight_layout()
```
